```python
import pandas as pd
output_base = 'mitf_regulon' 
import sys
import os

# Find repo root
if os.path.basename(os.getcwd()) == 'scripts':
    REPO_ROOT = os.path.dirname(os.getcwd())
else:
    REPO_ROOT = os.getcwd()

if REPO_ROOT not in sys.path: sys.path.append(REPO_ROOT)
from pan_cancer_config import ANALYSIS_SUFFIX
output_filename="mitf_metabolic_regulon_pairs"+ANALYSIS_SUFFIX+ ".csv"
# Load output
mitf_out = os.path.join(REPO_ROOT, "output", output_base, output_filename)
if os.path.exists(mitf_out):
    df = pd.read_csv(mitf_out)
    display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

if os.path.exists(mitf_out) and 'Metabolite_Name' in df.columns:
    plt.figure(figsize=(10,6))
    top_metabolites = df['Metabolite_Name'].value_counts().head(15)
    sns.barplot(y=top_metabolites.index, x=top_metabolites.values, palette='viridis')
    plt.title('Top 15 Metabolites Interacting with MITF Transcriptional Targets')
    plt.xlabel('Number of Target Genes')
    plt.ylabel('Metabolite')
    plt.show()
```
