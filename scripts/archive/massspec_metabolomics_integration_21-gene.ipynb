# Pan-Cancer Mass Spec Metabolomics Integration
    
### Goal
To integrate the 21-gene metabolic signature with the pan-cancer mass spectrometry metabolomics dataset (PMID: 29396322).

### Purpose
To show the association between the 21-gene and actual metabolite abundances across multiple cancer types. While the primary signature is derived from single-cell transcriptomics, verifying these metabolic axes at the metabolomic layer provides orthogonal validation.

### Key Insights
- **Metabolite Signature Score**: PCA-based scores of signature-linked metabolites show shifts between Tumor and Normal samples across the 7 major cancer types (KIRC, BLCA, BRCA, OV, PAAD, PRAD, LGG), where applicable.
- **Cross-Cohort Consistency**: The shifts in BRCA (Breast) and OV (Ovarian) metabolomic profiles align with our scRNA-seq findings, supporting the robustness of the metabolic network.
